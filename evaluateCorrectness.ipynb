```python
import pandas as pd
import hashlib

def canonical_edges(x):
    # x: [[a,b], [c,d], ...]
    if x is None or (isinstance(x, float) and pd.isna(x)):
        return None
    # make each edge unordered (min,max), then sort all edges
    edges = [tuple(sorted(e)) for e in x]
    edges.sort()
    return edges  # list of tuples

def hash_edges(edges):
    if edges is None:
        return None
    # stable bytes representation
    b = repr(edges).encode("utf-8")
    return hashlib.sha256(b).hexdigest()

results_h = resultsAlgo.copy()
combined_h = combinedData.copy()

results_h["graph_hash"] = results_h["graph"].apply(canonical_edges).apply(hash_edges)
combined_h["edges_hash"] = combined_h["edges"].apply(canonical_edges).apply(hash_edges)
combined_first = combined_h.drop_duplicates(subset="edges_hash", keep="first")

merged = results_h.merge(
    combined_first,
    how="inner",
    left_on="graph_hash",
    right_on="edges_hash",
    suffixes=("_results", "_combined")
)
```
