```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\fridm\Desktop\GraphPropertyPredict\results_8.csv')
```

```python
df8 = df[df['n'] == 8]
```

```python
# apply ln to time_ms
df8log = df8.copy()
df8log['time_ms'] = np.log(df8log['time_ms'])
df8log.hist(column='time_ms', bins=50)
```

```python
df8log = df8.copy()
df8log['time_ms'] = np.log(df8log['time_ms'])
df8log = df8log[df8log['result']==0]
df8log.hist(column='time_ms', bins=50)
```

```python
df8log = df8.copy()
df8log['time_ms'] = np.log(df8log['time_ms'])
df8log = df8log[df8log['result']==1]
df8log.hist(column='time_ms', bins=50)
```

```python
import matplotlib.pyplot as plt

# compute median time for each m
median_by_m = df8log.groupby('m')['time_ms'].median()

# scatterplot
plt.figure(figsize=(6, 5))
plt.scatter(median_by_m.index, median_by_m.values)

plt.xlabel('m')
plt.ylabel('Mediana tiempo (ms)')
plt.title('Mediana tiempo pra n=8 en funcion de aristas')
plt.grid(True)
plt.tight_layout()
plt.show()

```
